```python
import pandas as pd
import matplotlib as plt 
import seaborn as sns 
import numpy as np
from sklearn.preprocessing import LabelEncoder

df = pd.read_csv("../music_reccomendations/train.csv")
df_rec = pd.read_csv("../music_reccomendations/recommend.csv")
```

As we explored on the EDA, the column Age and Release date represent the same information we decide to drop the age column from the analysis due to music industry standards. Historically speaking the decades that the songs were recorded has major impact on the quality and  type of sound it takes leading us to prioritize specific dates as good references for music reccomendations.

```python
df = df.drop("age", axis=1).drop("Unnamed: 0", axis=1).drop("lyrics", axis=1).drop("track_name", axis=1)

df_rec = df_rec.drop(['age',"Unnamed: 0", "lyrics","track_name", "like/girls"], axis=1)
```

```python
labelencoder = LabelEncoder()

df['genre'] = labelencoder.fit_transform(df['genre'])
df['topic'] = labelencoder.fit_transform(df['topic'])
df['artist_name'] = labelencoder.fit_transform(df['artist_name'])
df
```

```python
df_rec['genre'] = labelencoder.fit_transform(df_rec['genre'])
df_rec['topic'] = labelencoder.fit_transform(df_rec['topic'])
df_rec['artist_name'] = labelencoder.fit_transform(df_rec['artist_name'])
```

```python
df.to_csv("Processed_data.csv", index=False)

df_rec.to_csv("Processed_data_rec.csv", index=False)
```
